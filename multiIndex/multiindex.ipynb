```python
pivoted = df.pivot_table(index=['Race', 'Character'],
                         columns='Film',
                         aggfunc='sum',
                         margins=True,
                         margins_name='All Films',
                         fill_value=0
                         ).sort_index()
pivoted.head(15)
```

```python
pivoted.index.names
```

```python
pivoted.sort_values(by=('Words', 'All Films'), ascending=False)
```

```python
order = [('Words', 'The Fellowship Of The Ring'),
         ('Words', 'The Two Towers'),
         ('Words', 'The Return Of The King'), 
         ('Words', 'All Films')]
pivoted.reindex(order, axis=1)
```
