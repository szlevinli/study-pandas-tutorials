```python
from pathlib import Path
import pandas as pd
pd.options.display.max_rows = 5
```

```python
data_dir = Path.cwd() / 'data'
org_file_name = '02_all_organizations.xlsx'
project_file_name = '04_cy_projects.xlsx'
```

```python
df = pd.read_excel(data_dir / org_file_name)
df
```

```python
# industry org id
industry_id = 1001
```

```python
# branch
df_branch = df.loc[df['pid'] == industry_id]
df_branch
```

```python
# project department
df_dept = pd.merge(
    df, df_branch[['id']], left_on='pid', right_on='id', suffixes=('', '_y'))
df_dept
```

```python
df_industry = pd.merge(df_branch, df_dept,
                       left_on='id', right_on='pid', suffixes=('_branch', '_dept'))
df_industry = df_industry[['id_branch',
                           'org_name_branch',
                           'id_dept',
                           'org_name_dept']]
df_industry
```

```python
df_project = pd.read_excel(data_dir / project_file_name)
df_project
```

```python
df_all = pd.merge(df_industry, df_project, left_on='id_dept', right_on='pid')
df_all = df_all[['id_branch', 'org_name_branch',
                 'id_dept', 'org_name_dept', 'org_id']]
df_all = df_all.rename(columns={'id_branch': 'branch_id',
                       'org_name_branch': 'branch_name',
                                'id_dept': 'dept_id',
                                'org_name_dept': 'dept_name',
                                'org_id': 'project_id'
                                })
df_all
```
