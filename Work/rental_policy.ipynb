```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
file_dir = Path.cwd() / 'data' / 'rental_policy'
file_path = sorted([f for f in file_dir.glob('20*.xlsx')])[-1]
file_path
```

```python
# 从 Excel 文件中读取数据
df = pd.read_excel(file_path)
# 解耦 Excel 文件中的字段
df = df.rename(columns={
  'ORGAN_ID': 'project_id',
  '项目名称': 'project',
  '项目公司': 'branch',
  '项目部': 'department',
  '资源编号': 'resource_no',
  '资源名称': 'resource_name',
  '业务状态': 'resource_status',
  '租赁状态': 'rental_status',
  '经营方式': 'mode_of_operation',
  '资源类型': 'resource_class',
  '租赁政策状态': 'rental_policy_status'
})
df
```

```python
# filter resource_status = '自用'
df = df[df['resource_status'] != '自用']
df
```

```python
# 根据租赁政策状态, 生成是否完成审核字段, 主要的作用是后面的代码写起来更简单
df['is_approval'] = df['rental_policy_status'].apply(lambda x: 0 if x == '未完成' else 1)
df
```

```python
# 根据 分公司, 项目部, 项目, 是否审核通过 来进行分组统计数量
df_count = df.groupby(['branch', 'department', 'project', 'is_approval'])['resource_no'] \
    .count() \
    .rename('project_is_count') \
    .reset_index()
df_count
```

```python
# 基于分公司和是否完成审核进行数量统计
branch_count = df_count.groupby(['branch', 'is_approval'])['project_is_count'] \
    .sum() \
    .rename('branch_is_count') \
    .reset_index()
branch_count
```

```python
# 根据分公司统计数量, 即不考虑是否完成审核
branch_count['branch_count'] = branch_count.groupby('branch')['branch_is_count'] \
    .transform('sum')
# 根据分公司计算完成和未完成审核的比率
branch_count['branch ratio'] = np.round((branch_count['branch_is_count'] /
                                         branch_count['branch_count']), 4)
# 计算整个产园已完成审核数据量
branch_count['approval_count'] = branch_count.loc[branch_count['is_approval']
                                                  == 1, 'branch_is_count'].sum()
# 计算整个产园未完成审核数量
branch_count['unapproval_count'] = branch_count.loc[branch_count['is_approval']
                                                    == 0, 'branch_is_count'].sum()
# 计算整个产园的数量
branch_count['total_count'] = branch_count['branch_is_count'].sum()
# 计算整个产园的完成审核比率, 可以看做是平均完成率
branch_count['approval average ratio'] = np.round(branch_count['approval_count'] /
                                                  branch_count['total_count'], 4)
# 计算整个产园的未完成审核比率, 可以看做是平局未完成率
branch_count['unapproval average ratio'] = np.round(branch_count['unapproval_count'] /
                                                    branch_count['total_count'], 4)
branch_count
```

```python
# TODO: 处理没有完成一单审核的分公司, 下面用的方式比较别扭, 以后想想有没有好的办法.
# branch_count['branch %'] == 1.0 用于处理一单都没审核完成数据
branch_analysis = branch_count.loc[(branch_count['is_approval'] == 1) | (
    branch_count['branch ratio'] == 1.0), ['branch', 'branch ratio', 'approval average ratio']]
# 对于一单都没审核完成的需要做取反操作
branch_analysis.loc[(branch_analysis['branch ratio'] == 1.0) & (
    branch_analysis['is_approval'] == 0), 'branch ratio'] = 0.0
branch_analysis
```

```python
output_file_path = Path.cwd() / 'output' / 'rental_policy' / f'{file_path.stem}_analysis.xlsx'
branch_analysis.to_excel(output_file_path)
```
